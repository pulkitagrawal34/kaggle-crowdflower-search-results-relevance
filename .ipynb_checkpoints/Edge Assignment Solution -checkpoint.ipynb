### EDA: Query and Product Title

```python
def count_common_words(sentA, sentB):
    
    if not sentA or not sentB:
        return 0 
    
    token_sentA = set(sentA.lower().split())    
    token_sentB = set(sentB.lower().split())
    
    common_word_count = 0 
    
    if token_sentA and token_sentB:
        
        for word in token_sentA:
            if word in token_sentB:
                common_word_count+=1
    
    
    return common_word_count
```

### Text Pre-processing

```python
def cleanHTML(phrase):
    soup = BeautifulSoup(phrase,'lxml')
    text = soup.get_text()
    return text
```

```python
df = pd.read_csv('train.csv')
df["product_description"].fillna('', inplace=True)
synonyms = pd.read_csv('synonyms.csv',names=['act','exp'])

actual = list(synonyms["act"].values)
changeto = list(synonyms["exp"].values)

replace_dict = pickle.load(open('replace.pkl','rb'))
```

```python
def word_replacement(text):
    
    for act,exp in zip(actual,changeto):
        text = re.sub(act,exp,text)
    for key,value in replace_dict.items():
        text = re.sub(str(key),str(value),text)
    return text
```

```python
def decontraction(phrase):
    phrase = phrase.lower()
    
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)
    phrase = re.sub(r"\’t|\'t", " not", phrase)
    phrase = re.sub(r"\’re|'re", " are", phrase)
    phrase = re.sub(r"’s|\'s", " is", phrase)
    phrase = re.sub(r"’d|\'d", " would", phrase)
    phrase = re.sub(r"’ll|\'ll", " will", phrase)
    phrase = re.sub(r"’ve|\'ve", " have", phrase)
    phrase = re.sub(r"’m|\'m", " am", phrase)
    
    return phrase
```

```python
stop_words = stopwords.words('english')
stop_words.remove('not')
stop_words.remove('no')
```

```python
token_pattern = r"\b\w\w+\b"

p = PorterStemmer()

def stemming(text):
    token = re.compile(token_pattern, flags = re.UNICODE)
    tokens = [x.lower() for x in token.findall(text)]
    out = []
    for token in tokens:
        out.append(p.stem(token))
    
    out = ' '.join([word for word in out])
    return out
```

```python
def text_preprocess(phrase, stem=False):
    
    phrase = re.sub(r'http/S+','',phrase)
    phrase = cleanHTML(phrase)
    phrase = word_replacement(phrase)
    
    if stem:
        phrase = stemming(phrase)
        
    phrase = decontraction(phrase)
    phrase = re.sub(r'[^0-9a-zA-Z ]+',' ',phrase)
    phrase = re.sub(r"\n", " ", phrase)    
    phrase = re.sub(' +', " ", phrase)
    
    words = phrase.split()
    output = []
    for word in words:
        if word not in stop_words and len(word)>1:
            output.append(word)
            
    output = ' '.join(word for word in output)
    
    return output
    
```

```python
#pre-processing query text
df['cleaned_query1'] = df.apply(lambda x : text_preprocess(str(x['query']).lower(),stem=True),axis = 1)
```

```python
# Replacing space with underscore to make it a single word.
df['cleaned_query'] = df.apply(lambda x : x['query'].replace(' ','_'),axis = 1)
```

### Feature Extraction

```python
def cosine_similarity(sentence1, sentence2):
    
    sentence1_embeddings = model.encode([sentence1])
    sentence2_embeddings =  model.encode([sentence2])

    cosine_similarity = np.dot(sentence1_embeddings, np.transpose(sentence2_embeddings))/(np.linalg.norm(sentence1_embeddings) * np.linalg.norm(sentence2_embeddings))
    
    return cosine_similarity[0][0]
```

```python
def create_bigrams(sentence):

    tokens = sentence.split()
    tokens = [word for word in tokens if word not in stop_words]
    return [" ".join(tokens[index: index+2]) for index in range(len(tokens)-1)]
```

```python
def common_bigrams(sentence1, sentence2):
    
    sent1_bigrams = create_bigrams(sentence1)
    sent2_bigrams = create_bigrams(sentence2)

    if len(sent1_bigrams) ==0 or len(sent2_bigrams) ==0:
        return 0
    
    counter = 0 

    for _b in sent1_bigrams:
        if _b in sent2_bigrams:
            counter+=1
    
    return counter/len(sent2_bigrams)
```

```python
def longest_common_sequence(s1, s2):
    matrix = [["" for x in range(len(s2))] for x in range(len(s1))]
    
    for i in range(len(s1)):
        for j in range(len(s2)):
            if s1[i] == s2[j]:
                if i == 0 or j == 0:
                    matrix[i][j] = s1[i]
                else:
                    matrix[i][j] = matrix[i-1][j-1] + s1[i]
            else:
                matrix[i][j] = max(matrix[i-1][j], matrix[i][j-1], key=len)

    cs = matrix[-1][-1]

    return len(cs)
```

```python
def plot_feature_importance(importance, names, model_type):

    feature_importance = np.array(importance)
    feature_names = np.array(names)

    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)

    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)


    plt.figure(figsize=(10,8))
    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])
    plt.title(model_type + ' feature importance')
    plt.xlabel('FEATURE IMPORTANCE')
    plt.ylabel('FEATURE NAMES')
```

```python
def permutation_importances(model, X_train, y_train, metric):
    y_pred = model.predict_proba(X_train)[:, 1]
    baseline = metric(y_train, y_pred)
    
    imp = []
    for col in X.columns:
        save = X_train[col].copy()
        X_train[col] = np.random.permutation(X_train[col])
        
        pred = model.predict_proba(X_train)[:, 1]
        m = metric(y_train, pred)
        X_train[col] = save
        imp.append(baseline - m)
        
    return np.array(imp)
```
